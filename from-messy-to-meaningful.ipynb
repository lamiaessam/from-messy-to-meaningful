```python
new_df = df.copy()
new_df['temperature'] = new_df['temperature'].fillna(new_df['temperature'].mean())
new_df['windspeed'] = new_df['windspeed'].fillna(new_df['windspeed'].mean())

new_df.loc[(new_df['temperature'] <10) & (new_df['event'].isnull()), 'event'] ='snow'
new_df.loc[(new_df['temperature'] >30) & (new_df['event'].isnull()), 'event']= 'sunny'
print("data frame with 'event' columns update based on temperature condations")
new_df
```
